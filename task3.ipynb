```python
import happybase
connection = happybase.Connection('localhost', port=9090)

table_name = 'Assignment_Table'
```

```python
t = connection.table(table_name)
list = []
id = "b'row_0'"
for key, data in t.scan():
    print(key, data)
    list.append(data[b'cf:clean_text'].decode('utf-8'))
    





```

```python
print(len(list))
```

```python
print(list)
```

```python
from textblob import TextBlob
score_list = []
label = []
for item in list:
    blob = TextBlob(item)
    sentiment_score = blob.sentiment.polarity
    score_list.append(sentiment_score)
    if sentiment_score > 0:
        label.append("Positive")
    elif sentiment_score < 0:
        label.append("Negative")
    else:
        label.append("Neutral")
```

```python
import pandas as pd
df = pd.DataFrame({'Sentences': list, 'Score': score_list,'Label':label})
```

```python
print(df)
```

```python
counts = df['Label'].value_counts(normalize=False)
print(counts)
```

```python
import numpy as np

df.to_csv('data.csv')
df = df.replace('', np.nan)

# drop rows with any NaN values
df = df.dropna()

print(df)
```
